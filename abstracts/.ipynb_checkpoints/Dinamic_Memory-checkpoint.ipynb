# Управление динамической памятью












---
- Stack: содерджит кадры функции;
- Heap: ипользуется для работы с динамической пямятью;
- Memory Mapping: динамические библиотеки, необходимые для запуска программы (libc.so), также является участком динамической памяти для работы с большими сегментами
- Вызовы нужно добовлять в обертку, например для sbrk()

```
char *ptr1;

ptr1 = sbrk(100);
if(ptr1 < 0) {
    perror("Out of mem");
    exit(EXIT_FAILURE);
}
```
- #include <malloc.h>: необходимо прописывать, чтобы работать с новой версией библиотеки, которая рассчитана на несколько потоков
- Всегда высвобождать память

---

### malloc()
- Принимает колличество байт, которое необходмо выделить
- Выделяет неициализированную область памяти

````c++
void *malloc(size_t size);
````

### calloc()
- Принимает колличество эллементов и размер одного эллемента
- Выделяет память и заполняет ее нулям

````c++
void *calloc(size_t nmemb, size_t size);
````

### free()
- Принимает адрес, который возвращают функции **malloc()** или **calloc()**
- Высвобаждает память

````c++
void free(void *ptr);
````

### realloc()
- Перевыделение размера памяти, как и уменьшает, так и увеличивает


````c++
void realloc(void *ptr, size_t size);
````

### alloca()
- Выделяет массив не в куче, а в стеке

````c++
void alloca(size_t size);
````
